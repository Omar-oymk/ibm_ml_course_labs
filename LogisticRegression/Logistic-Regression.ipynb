```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/ChurnData.csv"
df = pd.read_csv(url)
df.sample(5)
```

```python
# describe the data
df.describe()
```

```python
# show more info
df.info()
```

```python
# simplify it like just take the first 7 features blindly  (not optimal but we will do so for now to focus on practice)
df = df[["tenure", "age", "address", "income", "ed", "employ", "equip", "churn"]]
```

```python
# consider the correlation table
df.corr()
```

```python
# ill just keep the age and tenure since they are the highest corr features
df = df[["age", "tenure", "churn"]]
```

```python
# ill pick the features and labels and standardize the data
X = df[["age", "tenure"]]
y = df["churn"].astype(int)     # instead of float to make it represent a class
```

```python
# now ill plot the data
# scatter plot the values where the churn is 0
age_arr = X.age.values
tenure_arr = X.tenure.values
churn_arr = y.values

plt.scatter((age_arr)[churn_arr == 0], (tenure_arr)[churn_arr == 0], color = "red", label = "churn = 0")
plt.scatter((age_arr)[churn_arr == 1], (tenure_arr)[churn_arr == 1], color = "green", label = "churn = 1")
plt.title("AGE VS TENURE")
plt.xlabel("Age")
plt.ylabel("Tenure")
plt.legend()
plt.show()
```

```python
# now we will standardize the data using z-score standardization
from sklearn.preprocessing import StandardScaler
se = StandardScaler()
X_scaled = se.fit_transform(X)

# now ill split the data into testing and training sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state = 42)
```

```python
# we train the model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
# now we predict 
predictions = model.predict(X_test)
print(predictions)
```

```python
# we can now plot the results for each the true and the predictions to compare

plt.subplot(2, 1, 1)
plt.scatter((X_test[:, 0])[y_test.values == 0], (X_test[:, 1])[y_test.values == 0], color = "red", label = "churn = 0")
plt.scatter((X_test[:, 0])[y_test.values == 1], (X_test[:, 1])[y_test.values == 1], color = "green", label = "churn = 1")
plt.title("TRUE AGE VS TENURE (NOTE: these are scaled x values)")
plt.xlabel("Age")
plt.ylabel("Tenure")
plt.legend()


plt.subplot(2, 1, 2)
plt.scatter((X_test[:, 0])[predictions == 0], (X_test[:, 1])[predictions == 0], color = "red", label = "churn = 0")
plt.scatter((X_test[:, 0])[predictions == 1], (X_test[:, 1])[predictions == 1], color = "green", label = "churn = 1")
plt.title("PREDICTED AGE VS TENURE (NOTE: these are scaled x values)")
plt.xlabel("Age")
plt.ylabel("Tenure")
plt.legend()
plt.tight_layout(pad = 1)

plt.show()
```

```python
# so as we can see there are some errors but they are not that abundant
# this is a great visualization to show how to compare between true and predicted classification results
```

```python
# evaluate the model
from sklearn.metrics import log_loss, accuracy_score
print("The cross-entropy = ", round(log_loss(y_true= y_test, y_pred= predictions), 2))        # the less the better
print("The Accuracy = ", accuracy_score(y_true= y_test, y_pred= predictions))       # evaluates how many true predictions/ total predictions
```
